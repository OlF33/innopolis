```python
# функция разбора координат
def parse_coords(dic):
    import numpy as np
    
    # dic = eval(text)
    if dic['type'] == 'LineString':
        return np.array(dic['coordinates'])
    elif dic['type'] == 'Polygon':
        return np.array(dic['coordinates'][0])
    elif dic['type'] == 'MultiPolygon':
        lst = [np.array(x[0]) for x in dic['coordinates']]
        return np.concatenate(lst)
    elif dic['type'] == 'GeometryCollection':
        lst = [parse_coords(dic0) for dic0 in dic['geometries']]
        return np.concatenate(lst)
    
# вероятно, нет смысла особо вникать в северную широту и восточную долготу
# скорее всего, достаточно взять среднее из всех координат, представленных в описании
train['coord_nlat'] = 0
train['coord_elong'] = 0
for row in train.index:
    arr = parse_coords(eval(train['.geo'][row]))
    assert arr.ndim == 2, 'Должно быть ровно 2 измерения'
    assert arr.shape[1] == 2, 'Должно быть 2 столбца'
    avg = np.average(arr, axis=0)
    train.loc[row, 'coord_nlat'] = avg[0]
    train.loc[row, 'coord_elong'] = avg[1]
    
train[['.geo', 'coord_nlat', 'coord_elong']]
```

```python
# для проверки вычисления среднего достаточно одного
column_number = X.columns.searchsorted(X.iloc[0][train.iloc[0] == 0].index[0])
print(column_number, X.iloc[0, column_number])    # сразу убедимся, что выбрана нужная ячейка
# а также желательно убедиться, что соседние значения ненулевые, и сохранить значение, которое должно получиться
if X.iloc[0, column_number - 1] == 0 or X.iloc[0, column_number + 1] == 0:
    print('Не подходит')
else:
    mean_value = (X.iloc[0, column_number - 1] + X.iloc[0, column_number + 1])/2
    print(mean_value)
```

```python
# ещё надо проверить нулевые значения вегетационных индексов в крайнем левом и крайнем правом столбцах
left_row = X.iloc[:, 3][X.iloc[:, 3] == 0].index[0]
right_row = X.iloc[:, -1][X.iloc[:, -1] == 0].index[0]
print(X.iloc[left_row, 3], X.iloc[right_row, -1])
# аналогично предыдущему, проверим и запомним
if X.iloc[left_row, 4] == 0:
    print('Левый не подходит')
else:
    left_value = X.iloc[left_row, 4]
if X.iloc[right_row, -2] == 0:
    print('Правый не подходит')
else:
    right_value = X.iloc[right_row, -2]
```

```python
df_forward = X[X.columns[3:]].mask(X==0)
df_backward = df_forward.copy()
df_forward.fillna(method='ffill', axis=1, inplace=True)
df_forward.fillna(method='bfill', axis=1, inplace=True)    # для самых левых NaN'ов
df_backward.fillna(method='bfill', axis=1, inplace=True)
df_backward.fillna(method='ffill', axis=1, inplace=True)    # для самых правых

X.loc[:, X.columns[3:]] = (df_forward + df_backward)/2

assert X.isna().sum().sum() == 0, 'Не должно быть незаполненных ячеек'
assert (X == 0).sum().sum() == 0, 'Слишком много нолей'
assert X.iloc[0, column_number] == (X.iloc[0, column_number-1] + X.iloc[0, column_number+1])/2 and X.iloc[0, column_number] == mean_value, 'Среднее значение не равно  среднему'
assert X.iloc[left_row, 3] == X.iloc[left_row, 4] and X.iloc[left_row, 3] == left_value, 'Левое значение не равно следующему'
assert X.iloc[right_row, -1] == X.iloc[right_row, -2] and X.iloc[right_row, -1] == right_value, 'Правое значение не равно предыдущему'
print('Все тесты прошли благополучно')
```

Картина, полученная на диаграмме географического распределения сельскохозяйственных культур, позволяет предположить,
что нелинейные комбинации координат помогут надёжнее разделить культуры.<br>
Для проверки этой гипотезы построим модели на исходных данных и на данных, включающих степени координат.

```python
from sklearn.linear_model import RidgeClassifierCV
from sklearn.metrics import recall_score, make_scorer
from sklearn.preprocessing import StandardScaler

scorer = make_scorer(recall_score, average='macro')

X_poli = X[['coord_elong', 'coord_nlat']].rename(columns=dict(zip(['coord_elong', 'coord_nlat'], ['x', 'y'])))

cur_time = pd.Timestamp.now()
print('Начали:', cur_time.strftime('%X'))
max_degree = 10 
best_recall = 0
columns_old = X_poli.columns
for i in range(max_degree):
    # масштабируем факторы на всякий случай
    X_trans = StandardScaler().fit_transform(X_poli)
    rc = RidgeClassifierCV(alphas=np.arange(0.1, 10, 0.1), scoring=scorer, cv=5).fit(X_trans, train.crop)
    print(f'power {i+1}, recall {rc.best_score_}')
    # обновим, если надо, лучшую оценку
    if rc.best_score_ > best_recall:
        best_recall = rc.best_score_
        best_degree = i+1
    # дополним матрицу факторов очередной степенью
    if i < max_degree - 1:
        columns_new = []
        for column in columns_old:
            new_column = 'x'+column
            X_poli[new_column] = X_poli[column]*X_poli['x']
            columns_new.append(new_column)
        new_column = columns_old[-1]+'y'
        X_poli[new_column] = X_poli[columns_old[-1]]*X_poli['y']
        columns_new.append(new_column)
        columns_old = columns_new
    
new_time = pd.Timestamp.now()
print(f'Закончили: {new_time.strftime("%X")}, продолжительность {(new_time - cur_time).round("s")}')    
print(f'Лучшая оценка {best_recall} получена при степени {best_degree}')
del X_trans
```

Достаточно высокая оптимальная степень (10), скорее всего, говорит о том, что зависимость культуры от координат имеет более интересный характер, чем просто полиномиальный, - возможно, экспоненциальный, обратно пропорциональный, логарифмический... Вариантов слишком много, чтобы выбирать один конкретный (пришлось бы проверить все с применением нелинейной оптимизации), поэтому ограничимся добавлением новых переменных к нашему датасету.

```python
# для автоматизации подбора алгоритма
class ClfAlgorithm():
    def __init__(self, label, *, params={}, init_string=''):
        self.label = label                # для опознания
        self.params = params              # для GridSearchCV
        self.init_string = init_string    # для создания объекта (с помощью eval)
        
    def get_model(self):
        if self.init_string:
            return eval(self.init_string)
        else:
            return None

    # понадобится для настройки Pipeline через GridSearchCV
    def transform_params(self, prefix):
        return {prefix + '__' + item[0]: item[1] for item in self.params.items()}
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC, SVC
from sklearn.neighbors import KNeighborsClassifier, RadiusNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, ExtraTreeClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier, StackingClassifier
from xgboost import XGBClassifier
# для настройки и проверки понадобятся
from sklearn.model_selection import GridSearchCV, StratifiedKFold, train_test_split
# метрики
from sklearn.metrics import recall_score, make_scorer
```

```python
# для повторяемости результатов
random_state = 64

# заготовки моделей
log_regr = ClfAlgorithm(
    'LogisticRegression',
    params={'C': [0.05, 0.1, 0.15], 'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']},
    init_string='LogisticRegression(class_weight="balanced", max_iter=100000, random_state=random_state)'
)
lin_svc = ClfAlgorithm(
    'LinearSVC',
    params={'C': np.arange(0.05, 0.5, 0.05)},
    init_string='LinearSVC(class_weight="balanced", max_iter=500000, random_state=random_state)'
)
svc = ClfAlgorithm(
    'SVC',
    params={'C': np.arange(0.9, 1.5, 0.2), 'gamma': [0.09, 0.10, 0.11], 'coef0': np.arange(-1.5, 1.1, 0.5)},
    init_string='SVC(class_weight="balanced", decision_function_shape="ovr", kernel="rbf")'
)
kneigh = ClfAlgorithm(
    'KNeighborsClassifier',
    params={'n_neighbors': range(1, 50, 2), 'weights': ['uniform', 'distance']},
    init_string='KNeighborsClassifier()'
)
rneigh = ClfAlgorithm(
    'RadiusNeighborsClassifier',
    params={'radius': np.arange(0.5, 50, 0.5), 'weights': ['uniform', 'distance']},
    init_string='RadiusNeighborsClassifier(outlier_label=0)'
)
dec_tree = ClfAlgorithm(
    'DecisionTreeClassifier',
    params={'min_samples_split': range(10, 30, 4), 'min_samples_leaf': range(1, 20, 4)},
    init_string='DecisionTreeClassifier(random_state=random_state)'
)
extra_tree = ClfAlgorithm(
    'ExtraTreeClassifier',
    params={'min_samples_split': range(2, 103, 2), 'min_samples_leaf': range(1, 102, 2), 'max_features': [0.4, 0.6, 0.8, 1.0]},
    init_string='ExtraTreeClassifier(random_state=random_state)'
)
ada_boost = ClfAlgorithm(
    'AdaBoostClassifier',
    params={'n_estimators': range(5, 86, 10), 'learning_rate': np.arange(0.2, 0.8, 0.1)},
    init_string='AdaBoostClassifier(random_state=random_state)'
)
grad_boost = ClfAlgorithm(
    'GradientBoostingClassifier',
    params={'learning_rate': [0.15, 0.2, 0.25], 'n_estimators': [140, 160, 180]},
    init_string='GradientBoostingClassifier(random_state=random_state)'
)
bagging = ClfAlgorithm(
    'BaggingClassifier',
    params={'n_estimators': [5, 8, 10, 14, 20], 'max_features': [0.2, 0.4, 0.6, 0.8]},
    init_string='BaggingClassifier(random_state=random_state)'
)
xgb = ClfAlgorithm(
    'XGBClassifier',
    params={'n_estimators': [35, 40, 45, 70], 'subsample': np.arange(0.5, 1.0, 0.2)},
    init_string='XGBClassifier(eval_metric=recall_score, objective="multi:softprob", random_state=random_state)'
)

stack = ClfAlgorithm(
    'StackingClassifier'    # будет обучаться вне цикла
)

# для удобства обработки в цикле
clfs = [log_regr, lin_svc, svc, kneigh, rneigh, dec_tree, extra_tree, ada_boost, grad_boost, bagging, xgb, stack]
labels = [clf.label for clf in clfs]
```

```python
# алгоритмы градиентного спуска и некоторые другие требуют сравнимых по амплитуде исходных данных
# то есть нужна нормализация
from sklearn.preprocessing import StandardScaler
# а чтобы полноценно воспользоваться нормализацией, надо ещё Pipeline,
from sklearn.pipeline import Pipeline
# а также кэш и средство его очистки
from tempfile import mkdtemp
from shutil import rmtree

# для GridSearchCV
scorer = make_scorer(recall_score, average='macro')
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)

# кэш
cachedir = mkdtemp()

# для ускорения процесса
n_jobs = 10

# соберём, а потом выведем информацию о параметрах, чтобы быть уверенными, что нигде оптимизатор не упёрся в границы
# хотя это не гарантия нахождения оптимальных значений, но всё же некоторая подсказка
params_store = {label: [] for label in labels}
# для хранения результатов
df_models = pd.Series(data=np.empty((len(labels), ), dtype=object), index=labels)
df_recall = pd.Series(data=np.empty((len(labels), ), dtype=float), index=labels)

# предполагается длительная операция, нам интересно, сколько времени уйдёт на каждый этап
start_time = pd.Timestamp.now()
cur_time = start_time
print('Начали', cur_time.strftime('%X'))

# проверяем все модели, потом сравним
X_train, X_test, y_train, y_test = train_test_split(X, train.crop, test_size=0.2, stratify=train.crop, random_state=random_state)
for clf in clfs[:-1]:
    model = Pipeline([('scaler', StandardScaler()), ('clf', clf.get_model())], memory=cachedir)
    params = clf.transform_params('clf')

    gs = GridSearchCV(model, params, scoring=scorer, cv=skf, n_jobs=n_jobs)    
    gs.fit(X_train, y_train)
    df_models.loc[clf.label] = gs.best_estimator_
    params_store[clf.label].append(gs.best_params_)

    y_pred = gs.best_estimator_.predict(X_test)
    recall = recall_score(y_test, y_pred, average='macro')
    df_recall.loc[clf.label] = recall
    
    # выведем информацию об алгоритме и время
    new_time = pd.Timestamp.now()
    print(f'  algorithm: {clf.label}, recall: {recall:.3f}, {new_time.strftime("%X")}, duration {(new_time-cur_time).round("s")}')
    cur_time = new_time
# выберем лучшие модели
best_metrics = df_recall[df_recall > 0.85]
# список моделей для стэкинга
estimators = [(label, model) for label, model in zip(best_metrics.index, df_models[best_metrics.index])]
# для настройки гиперпараметров мета-алгоритм обернём в GridSearchCV
gs = GridSearchCV(XGBClassifier(eval_metric=recall_score, objective='multi:softprob', random_state=random_state),
                  {'n_estimators': [35, 40, 45, 70], 'subsample': np.arange(0.5, 1.0, 0.2)},
                  scoring=scorer, cv=skf)
model = StackingClassifier(estimators=estimators, final_estimator=gs, cv=skf).fit(X_train, y_train)    
label = 'StackingClassifier'
y_pred = model.predict(X_test)

df_models.loc[label] = model
params_store[label].append(model.final_estimator_.best_params_)
recall = recall_score(y_test, y_pred, average='macro')
df_recall.loc[clf.label] = recall

new_time = pd.Timestamp.now()
print(f'  algorithm: {label}, recall: {recall:.3f}, {pd.Timestamp.now().strftime("%X")}, duration {(new_time-cur_time).round("s")}')
    
stop_time = pd.Timestamp.now()
print('Готово,', stop_time.strftime("%X"), (stop_time - start_time).round('s'))
rmtree(cachedir)
```
